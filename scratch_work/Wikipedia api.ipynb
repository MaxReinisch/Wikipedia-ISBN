```python
import wikipedia
import re
```

```python
isbn_re1 = re.compile('(ISBN[-]*(1[03])*[ ]*(: ){0,1})*(([0-9Xx][- ]*){13}|([0-9Xx][- ]*){10})')
```

```python
isbn_re2 = re.compile('ISBN')
```

```python
easter_island = wikipedia.page("Easter_Island")
easter_island.url
```

```python
easter_island.content
```

```python
easter_island.links
```

```python
found = isbn_re1.findall(easter_island.content)
```
